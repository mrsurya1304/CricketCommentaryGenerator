Imorting libraries

```python
import pandas as pd
```

Reading the datasets (IPL commentery of 3 years)

```python
data1=pd.read_csv("/content/drive/MyDrive/NLP_CommentaryDatasets/IPL2018 - CommentaryData.csv")
data2=pd.read_csv("/content/drive/MyDrive/NLP_CommentaryDatasets/IPL2019 - CommentaryData.csv")
data3=pd.read_csv("/content/drive/MyDrive/NLP_CommentaryDatasets/IPL2020 - CommentaryData.csv")
```

```python
data1
```

Reversing the order and resetting the index to get commentary of 1st ball 1st match to the end of the tournament

```python
data1 = data1.sort_index(axis=0 ,ascending=False)
data1.reset_index(inplace = True, drop = True)
data1.head()
```

Renaming columns

```python
data1.rename(columns = {'short_text':'prompt','long_text':'commentary'}, inplace = True)
data1.head()
```

Seperating batsman to bowler and the event that occurred

```python
result = data1['prompt'].str.split(', ',n=1,expand=True)
data1["input"] = result[0]
data1["commentary"] = result[1]+", "+data1["commentary"]+"."
data1["event"] = result[1]
data1.head()
```

Retaining the required rows and making them to string inputs

```python
data1=data1[["input","event","commentary"]]
data1["event"] = data1["event"].str.strip("")
data1["commentary"] = data1["commentary"].str.strip("")
data1.head()
```

Function to carry out the previous sequence of steps to all the datasets

```python
def preparedata(data1):
  data1 = data1.sort_index(axis=0 ,ascending=False)
  data1.reset_index(inplace = True, drop = True)
  data1.rename(columns = {'short_text':'prompt','long_text':'commentary'}, inplace = True)
  result = data1['prompt'].str.split(', ',n=1,expand=True)
  data1["input"] = result[0]
  data1["commentary"] = result[1]+", "+data1["commentary"]+"."
  data1["event"] = result[1]
  data1["event"] = data1["event"].str.strip("")
  data1["commentary"] = data1["commentary"].str.strip("")
  data1=data1[["input","event","commentary"]]
  return data1
```

Preparing the data for all the datasets

```python
data2 = preparedata(data2)
data3 = preparedata(data3)
data2.head()
```

Concatenating all the datasets to a single dataset, resetting the index and converting to a dataframe

```python
commentarydata = [data1,data2,data3]
commentary = pd.concat(commentarydata)
commentary.reset_index(inplace = True, drop = True)
commentary = pd.DataFrame(commentary)
commentary.head()
```

Splitting batsman and bowler names and appending to the data frame

```python
result = commentary["input"].str.split(" to ",n=1)
batsman=[]
bowler=[]
for item in result:
  bowler.append(item[0])
  if(len(item)>1):
    batsman.append(item[1])
  elif(len(item)==1):
    batsman.append("missing")
```

```python
commentary["bowler"] = bowler
commentary["batsman"]= batsman
commentary.head()
```

Using these splitted names to remove these proper noun names in the commentary column (To attempt to provide generic commentary)

```python
commentary["commentary"]=commentary.apply(lambda x: str(x['commentary']).replace(str(x['bowler']),'the bowler'),axis=1)
commentary["commentary"]=commentary.apply(lambda x: str(x['commentary']).replace(str(x['batsman']),'the batsman'),axis=1)
```

```python
commentary.head()
```

Spelling correction for the commentary column to remove typos

```python
from textblob import TextBlob
commentary["commentary"].apply(lambda txt: ''.join(TextBlob(str(txt)).correct()))
```

Splitting the data (90-10) to get our training and testing data

```python
from sklearn.model_selection import train_test_split
x = commentary["event"]
y= commentary["commentary"]
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1)
```

Altering the train and test data to suit the format for GPT language generation models (GPT-2)

```python
traindata = pd.DataFrame(ytrain)
testdata = pd.DataFrame(xtest)
traindata["commentary"] = traindata["commentary"].str.strip("")
testdata["event"] = testdata["event"].str.strip("")
```

Resetting the index for our training and testing data

```python
traindata.reset_index(inplace = True, drop = True)
testdata.reset_index(inplace = True, drop = True)
```

Save the data onto Google Drive

```python
traindata.to_csv("/content/drive/MyDrive/NLP_CommentaryDatasets/T5train.csv",index=False)
testdata.to_csv("/content/drive/MyDrive/NLP_CommentaryDatasets/T5test.csv",index=False)
```
